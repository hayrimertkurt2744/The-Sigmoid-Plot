```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z): #sigmoid function
  return 1.0/(1.0+np.exp(-z))
```

```python
# Define the cost function J(ϕ(z), y; w)
def cost_function(phi_z, y):
    return -y * np.log(phi_z) - (1 - y) * np.log(1 - phi_z)
```

```python
z=np.arange(-10,10,0.1)
phi_z=sigmoid(z)
```

```python
plt.plot(z,phi_z,color="green")
plt.axvline(0.0, color="k")
plt.axhspan(0.0, 1.0, facecolor='1.0', alpha=1.0, ls='dotted')
plt.axhline(y=0.5, ls='dotted', color='k')
plt.yticks([0.0, 0.5, 1.0])
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (z)$')
plt.show()
```

```python
phi_z_values = np.linspace(0.01, 0.99, 100)
plt.figure(figsize=(8, 6))
plt.plot(phi_z_values, cost_function(phi_z_values, y=1), label='J(ϕ(z), y=1)')
plt.plot(phi_z_values, cost_function(phi_z_values, y=0), label='J(ϕ(z), y=0)')
plt.xlabel('ϕ(z) - Predicted Probability')
plt.ylabel('Cost')
plt.title('Logistic Regression Cost Function')
plt.legend()
plt.show()
```
